## CTR Prediction Model Training

### Main Arguments
Arguments that should be specified. These arguments determine the model/data (path)/task setting.

``` yaml
model_name: DIEN
dataset: MovieLens_1mimpCTRAll
path: '/work/lihanyu/ReChorus2/data'
num_neg: 0
metric: AUC,Log_loss
include_item_features: 1 
include_situation_features: 1 
model_mode: CTR
loss_n: BCE
```

`DIEN` is a sequential context-aware model. Context `include_item_features` and `include_situation_features`.

The `model_mode` for context-aware models includes `["CTR", "TopK"]` depending on the task.

For CTR task the `metrics` should be AUC and Log_loss, and `num_neg` (negative sample num) should be 0 when not performing negative sampling in training.

The `loss_n` for context-aware models includes `["BCE","BPR"]`, see in [BaseContextModel.py](https://github.com/THUwangcy/ReChorus/tree/master/src/models/BaseContextModel.py)

````python
!python /work/lihanyu/ReChorus/src/main.py --model_name DIEN --lr 5e-3 --l2 1e-6 --history_max 20 --alpha_aux 0.5 --aux_hidden_layers "[64,64,64]" --fcn_hidden_layers "[256]" --evolving_gru_type AIGRU --dropout 0.2 --dataset MovieLens_1mimpCTRAll --path '/work/lihanyu/ReChorus2/data' --num_neg 0 --batch_size 1024 --metric AUC,Log_loss --include_item_features 1 --include_situation_features 1 --model_mode CTR --loss_n BCE
````
