## Context-aware Top-K Recommendation Training

### Main Arguments
Arguments that should be specified. These arguments determine the model/data (path)/task setting.

``` yaml
model_name: DIEN
dataset: MovieLens_1mRankingAll
path: '/work/lihanyu/ReChorus2/data'
num_neg: 1
metric: NDCG,HR
include_item_features: 1 
include_situation_features: 1 
model_mode: TopK
```

`DIEN` is a sequential context-aware model. Context `include_item_features` and `include_situation_features`.

The `model_mode` for context-aware models includes `["CTR", "TopK"]` depending on the task.

For Top-K task the `metrics` should be NDCG and HR, and `num_neg` (negative sample num) should not be 0. 

No `loss_n` should be specified, because only BPR is supported in the topK task.

````python
!python /work/lihanyu/ReChorus/src/main.py --model_name DIEN --lr 5e-4 --l2 1e-6 --history_max 20 --alpha_aux 0.1 --aux_hidden_layers "[64]" --fcn_hidden_layers "[64]" --evolving_gru_type AIGRU --dropout 0 --dataset MovieLens_1mRankingAll --path '/work/lihanyu/ReChorus2/data' --num_neg 1 --batch_size 256 --eval_batch_size 32 --metric NDCG,HR --topk 3,5,10,20 --include_item_features 1 --include_situation_features 1 --model_mode TopK
````
